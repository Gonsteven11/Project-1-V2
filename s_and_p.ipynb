```python
data_SP500_Ratio = []
data_SP500 = []
for region in tables:
  df_tmp = nasdaqdatalink.get(region, paginate=True)
  df_tmp.columns = [region.split('/')[-1]]
  if '_PBV_RATIO_YEAR' in region:
    data_SP500_Ratio.append(df_tmp)
  else:
    data_SP500.append(df_tmp)
```

```python
pd.concat(data_SP500_Ratio).sort_index().ffill().plot(figsize=(13, 10), subplots=True)
```
